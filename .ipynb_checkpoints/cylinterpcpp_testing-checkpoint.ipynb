```python
import cylinterp
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

import matplotlib.pyplot as plt
plt.style.use('default')
```

## Diffusion Results
This sets up the drift which allows us to compute drift tiems

```python
tracks = np.fromfile('/data/jianyang/simulations/s2_electron_drift/tracks_0.bin', dtype = 'float64').reshape(1000,3000,3)
ztracks = tracks[:,:,0]
rtracks = tracks[:,:,1]
thtracks = tracks[:,:,2]
status = np.fromfile('/data/jianyang/simulations/s2_electron_drift/status_0.bin', dtype = 'int32')
init_points = np.fromfile('/data/jianyang/simulations/s2_electron_drift/init_pos_0.bin', dtype = 'float64').reshape(10,3)
```

```python
len(rtracks)
```

### Plotting (x,y) tracks

```python
radius_of_trouble = 200e-4
pangle = np.linspace(0,2*np.pi, 100)
cart_points = cylinterp.Tools.to_cartesian(init_points)

plt.figure(figsize = (10,10))

for i in range(10):
    plt.scatter(cart_points[i][0], cart_points[i][1], s = 5, color = 'g')

for i in range(1000):
    if status[i]==0:
        c = 'blue'
    elif status[i]==1:
        c = 'red'
    elif status[i]==2:
        c = 'black'
    else:
        c = 'green'
    buffer_cut = rtracks[i]!=0
    plt.plot(rtracks[i][buffer_cut]*np.cos(thtracks[i][buffer_cut]),
             rtracks[i][buffer_cut]*np.sin(thtracks[i][buffer_cut]), color = c, linewidth = 0.25)
plt.xlim(-2.5,2.5)
plt.ylim(-2.5,2.5)
plt.xlabel('x [cm]', fontsize = 15)
plt.ylabel('y [cm]', fontsize = 15)
plt.show()
```

### Plotting (r^2,z) tracks

```python
plt.figure(figsize = (5,10))

for i in range(10):
    plt.scatter(init_points[i][1]**2, init_points[i][0], s = 5, color = 'g')

for i in range(1000):
    if status[i]==0:
        c = 'blue'
    elif status[i]==1:
        c = 'red'
    elif status[i]==2:
        c = 'black'
    else:
        c = 'green'
    buffer_cut = rtracks[i]!=0
    
    plt.plot(rtracks[i][buffer_cut]**2,
             ztracks[i][buffer_cut], color = c, linewidth = 0.25)

central_z = 2.465952715464704
plt.hlines([5+central_z/2, 5-central_z/2], 0,2.5**2, linestyle = 'dashed', color = 'r')
plt.xlabel(r'r$^2$ [cm$^2$]', fontsize = 25)
plt.ylabel('z [cm]', fontsize = 25)
plt.show()
```
